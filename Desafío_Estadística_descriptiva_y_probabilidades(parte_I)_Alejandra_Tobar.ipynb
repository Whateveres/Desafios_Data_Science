```python
import pandas as pd

data = pd.read_csv("ds_salaries.csv")
data.head()
```

**1. El promedio general de los salarios, desviaci칩n est치ndar, quintiles y rango. (hint:escoge la columna adecuada para comparar**



```python
#Promedio general de los salarios en USD
promedio = data['salary_in_usd'].mean().round(2)
print(f"El Promedio es de {promedio}")

#Calculando Desviaci칩n Est치ndar
desv_std = data['salary_in_usd'].std().round(2)
print(f"La Desviaci칩n Est치ndar es {desv_std}")


#Calculando Quintiles
quintiles = data['salary_in_usd'].quantile([0.2,0.4,0.6,0.8]).round(2)
print(f"Los quintiles son:\n{quintiles}")


#Calculando Rango
minimo = data['salary_in_usd'].min().round(2)
maximo = data['salary_in_usd'].max().round(2)
rango = maximo - minimo
print(f"El Rango es {rango}")
```

**2.Una comparaci칩n entre los salarios agrupados por 3 diferentes categor칤as a tu elecci칩n. 쯇ara cu치l de ellas las medidas de tendencia central son m치s
representativas? 쯇ara cu치l son menos representativas? Explica**



```python
#Comparando salarios agrupados por categor칤as: job_title, experience_level, employment_type.

#Agrupando por categor칤a job_title:
cat_job_title = data.groupby('job_title')['salary_in_usd'].agg(['mean', 'median', 'std']).sort_values(by='mean', ascending=False).round(2)
print(cat_job_title.head())
print("\n")  # agrega un salto de espacio

#Agrupando por categor칤a experience_level:
cat_experience_level = data.groupby('experience_level')['salary_in_usd'].agg(['mean', 'median','std']).sort_values(by='mean', ascending=False).round(2)
print(cat_experience_level.head())
print("\n")  # agrega un salto de espacio

#Agrupando por categor칤a employment_type:
cat_employment_type = data.groupby('employment_type')['salary_in_usd'].agg(['mean', 'median','std']).sort_values(by='mean', ascending=False).round(2)
print(cat_employment_type.head())
print("\n")  # agrega un salto de espacio


```

Analizando las medidas de tendencia central como la Media y Mediana junto con la medida de dispersi칩n Desviaci칩n Est치ndar para el estudio de la distribuci칩n de los salarios por categor칤a, se observa que:

*  **Job_title**: Los datos podr칤an sugerir una distribuci칩n sim칠trica en torno a la media, ya que la distancia entre la media y la mediana es peque침a, indicando una distribuci칩n uniforme.  Sin embargo, la falta de datos suficientes para ciertos roles y la alta variabilidad mostrada en las desviaciones est치ndar en los salarios para algunos roles hacen que esta observaci칩n sea incierta.

* **Experience_level:** Esta categor칤a muestra comportamiento intermedio en comparaci칩n con las otras dos junto con variabilidad moderada reflejada en las desviaciones est치ndar. Sugiere una distribuci칩n m치s uniforme de los salarios dentro de cada nivel de experiencia, aunque con un poco menos simetr칤a que la categor칤a job_title.

*  **Employment_type:** La mayor distancia entre la media y mediana indica una posible presencia de outliers, ya que la media es m치s sensible a valores at칤picos. Esta hip칩tesis se refuerza por la mayor variabilidad en los salarios mostrada por las desviaciones est치ndar.


*Teniendo lo anterior en consideraci칩n, se concluye que las medidas de tendencia central son m치s representativas para las 3 categor칤as en el siguiente orden:*

1.   Experience_level
2.   Employment_type

**3. 쯈u칠 cargos reciben mejores sueldos en las empresas con sede en Estados Unidos?**



```python
#Filtrando por empresas con sede en US
data_us = data[data['company_location'] == 'US']

#Filtrando por cargos que reciben mejores sueldos
mejores_sueldos = data_us.groupby('job_title')['salary_in_usd'].max()

#Ordenando resultados de forma descendente
mejores_sueldos = mejores_sueldos.sort_values(ascending=False)

#Imprimiendo resultados
print(f"Los cargos que reciben mejores sueldos en las empresas con sede en Estados Unidos son:\n{mejores_sueldos.head()}")
```
